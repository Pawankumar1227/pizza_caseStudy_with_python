```python
import numpy as np
import pandas as pd
```

```python
#import all File into Google colab

detail = pd.read_csv('/content/order_details.csv')
ord= pd.read_csv('/content/orders.csv')
types = pd.read_csv('/content/pizza_types.csv',encoding='latin1')
pizza = pd.read_csv('/content/pizzas.csv')
```

```python
#create Join details and orders Tables

df_1 = detail.merge(ord , how = 'inner' , on = 'order_id')
df_1.head()
```

```python
# Create join pizza_type and pizza Tables 

df_2 =types.merge(pizza , how = 'inner' , on = 'pizza_type_id')
df_2.head()
```

```python
#Geather All Tables into One table

df = df_1.merge(df_2, how = "inner" , on = 'pizza_id')
df.head(2)
```

```python
#Split Date Coulmn to get Day

df['day']=df['date'].str.split('-').str[0]
df.head(2)
```

```python
#split TIme Column to get HOur

df['hour']=df['time'].str.split(':').str[0]
df.head(2)
```

```python
df.info()
```

```python
#Convert day and Hour Column Into int Datatype 

df['day'] = df['day'].astype('int64').inplace = True
df['hour'] = df['hour'].astype('int').inplace = True
```

```python
#Chacking Null value Column have

df.isnull().sum()
```

```python
#Retrieve the total number of orders placed.

df['order_id'].count()
```

```python
# Add a New column Revenue

df['revenue'] = df['price'] * df['quantity']
df.head(2)
```

```python
#Calculate the total revenue generated from pizza sales.

df['revenue'].sum()
```

```python
#Identify the highest-priced pizza.

df[['name','price']].max(axis=0)
```

```python
#Identify the most common pizza size ordered.


df.groupby('size')['size'].count().sort_values(ascending =False).head(1)
```

```python
#List the top 5 most ordered pizza types along with their quantities.

df.groupby('name')['name'].count().sort_values(ascending = False).head(5)
```

```python
#Find the total quantity of each pizza category ordered
# (this will help us to understand the category which customers prefer the most).


df.groupby('category')['category'].count().sort_values(ascending = False).head(1)
```

```python
#Determine the distribution of orders by hour of the day (at which time the orders are maximum
# (for inventory management and resource allocation)


df.groupby('hour')['hour'].count()
```

```python
#Find the category-wise distribution of pizzas (to understand customer behaviour).

df.groupby('category')['category'].count()
```

```python
#Determine the top 3 most ordered pizza types based on revenue
#(let's see the revenue wise pizza orders to understand from sales perspective which pizza is the best selling)

df.groupby('name')['revenue'].sum().sort_values(ascending = False).head(3)
```

```python
#Calculate the percentage contribution of each pizza type to total revenue
# (to understand % of contribution of each pizza in the total revenue)

df.groupby('name')['revenue'].sum()/df['revenue'].sum()*100
```
